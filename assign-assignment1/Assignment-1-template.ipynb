This function will take two parameters:
- the number of people in the household (an integer)
- a list of purchases, e.g. `[5.0 1 10.0 0 6.77 2 3.0 1]` , a concatenation of `float` and `integer`, where each expense (positive `float` number) has to be followed by the ID of the person who made the purchase (`integer`), which needs to be in the right range (greater or equal to 0 and less than the number of people in the household)

Otherwise (if the input does not satisfie the requirements), an error will be printed and an **empty list** will be returned.

````python
#// BEGIN_TODO [get_individual_totals] Get individual totals 
from typing import Union


def get_individual_totals(amount: int, purchases: (list[Union[float, int]])) -> list:
    """Return a list of individual totals of all the individual purchases.
    :param amount: the number of people in the household
    :param purchases: a list of purchases followed by ID's of the person
    :type amount: int
    :type purchases: list consisting of float followed by int
    :rtype: list
    :return: a list of individual totals of all people in the household


    >>> get_individual_totals(5, [5.0, 1, 1.2, 2, 6.77, 0, 3.0, 2])
    [5.0, 4.2, 6.77, 0.0, 0.0]

    >>> get_individual_totals(2, [15.00, 0, 23.15, 0])
    [38.15, 0.0]

    >>> get_individual_totals(2, [0.0, 0.0])
    Id's should be integer, please check your input.

    >>> get_individual_totals(2, [0.0, 1, 2.0])
    Not every cost has an associated ID, please double-check every cost has an ID in the input list.
    
    """

    if len(purchases) % 2 != 0:
        return "Not every cost has an associated ID, please double-check every cost has an ID in the input list."
        
    individual_totals = {}
    for i in range(1, len(purchases), 2): 
        if type(purchases[i-1]) is not float:
            return "Costs should be float, please check your input."
        if type(purchases[i]) is not int:
            return "Id's should be integer, please check your input."
        if purchases[i] in individual_totals:
            individual_totals[purchases[i]] += purchases[i - 1]
        else: 
            individual_totals[purchases[i]] = purchases[i - 1]

    for i in range(amount):
        if i not in individual_totals:
            individual_totals[i] = 0.0

    sorted_totals = sorted(individual_totals.items())
    return [total for id, total in sorted_totals]

#// END_TODO [get_individual_totals]

get_individual_totals(2, [15.00, 0, 23.15, 0])
````

## Task 2: Final Balances

We then create a second function that 
- takes the total spent by each individual: a list of floats, where the float number at index `i` corresponds to the total spent by the person with ID `i`
- returns the final balance of each individual: a list of floats, where the float number at index `i` corresponds to the final balance of the person with ID `i`

**Notes**:
* The final balance of an individual can be calculated by doing `spent - avg`, where
    * `spent` is the total amount spent by the individual
    * `avg` is the total amount spent by the group, divided by the number of people in the group
* Remember to deal with the possibily that the input list is empty
* Do not forget to write the **type hints** and **docstring** of the function.
* Include at least _two_ (2) `doctest` examples

**Examples**:

*Input:*  
```python
get_individual_final_balances([10.0, 20.0, 3.0])
```

*Output:*  

```python
[-1.0, 9.0, -8.0]
```

````python
#// BEGIN_TODO [get_individual_final_balances] Get individual final balances

def get_individual_final_balances(individual_totals: list) -> list:
    """Determines the final balances of each individual based on the total spent by this individual
    
    :param individual_totals: A list of the total spent by each individual where each index in the list represents a person and its value the total spent
    :type individual_totals: list
    :rtype: list of floats
    :return: A list of the final balances spent by each individuals

    >>> get_individual_final_balances([10.0, 20.0, 3.0])
    [-1.0, 9.0, -8.0]

    >>> get_individual_final_balances([])
    Please enter an input list

    """
    final_balances = []

    if len(individual_totals) == 0:
        return 'Please enter an input list'

    avg = sum(individual_totals) / len(individual_totals)

    for spent in individual_totals:
        final_balance = spent - avg
        final_balances.append(final_balance)

    return final_balances

#// END_TODO [get_individual_final_balances]

get_individual_final_balances([10.0, 20.0, 3.0])
````

````python
#// BEGIN_TODO [simplified_splitser] Create wrapper function for Simplified Splitser

def simplified_splitser(amount: int, purchases: (list[Union[float, int]])) -> list:
    """ Takes the number of people and a list of purchases as input and returns the final balances as a list

    :param amount: the number of people in the household
    :param purchases: a list of purchases followed by ID's of the person
    :type amount: int
    :type purchases: list consisting of float followed by int
    :rtype: list
    :return: a list of the final balances of every person with the index as ID

    >>> simplified_splitser(5, [12.45, 1, 15.6, 4, 25.5, 2, 4.30, 1])
    [-11.57, 5.18, 13.93, -11.57,  4.029999999999999]

    >>> simplified_splitser(5, [12.45, 1, 107.955, 4])
    [-24.081, -11.631, -24.081, -24.081, 83.874]

    >>> get_individual_totals(2, [0.0, 0.0])
    Id's should be integer, please check your input.

    >>> get_individual_totals(2, [0.0, 1, 2.0])
    Not every cost has an associated ID, please double-check every cost has an ID in the input list.
    
    """
    return get_individual_final_balances(get_individual_totals(amount, purchases))

#// END_TODO [simplified_splitser]

simplified_splitser(5, [12.45, 1, 15.6, 4, 25.5, 2, 4.30, 1])
````

### 3.2: Data Conversion
We now create another function that instead of two parameters only takes 1 parameter: 
- a list of strings. 

This means that the number of people and their purchases will be merged into one list of strings and it is our job to figure out how to
- transform the list into parameters that are fit for `simplified_splitser`
- **print** the output of `simplified_splitser` as multiple lines with 
    - each line having format "[name]: € [cost]" and 
    - the cost having **2 decimal points** (.2f).

**Notes:** 

- To check if a string `my_string` can be converted to an int, you can use `my_string.isdigit()`, this outputs a boolean. 
- To check if a string `my_string` can be converted to a float, you can use `my_string.replace('.', '', 1).isdigit()`, this outputs a boolean. 
- To convert a string `my_string` into a float, you can use the function `float(my_string)`
* Do not forget to write the **type hints** and **docstring** of the function.
* Include at least _three_ (3) `doctest` examples
    * Two with an input that can be correctly converted
    * One with input that will not following the requirements when converted
        * This means that errors from functions that are called will be printed
            
**Examples:**

*Input:* 
```python

    ["5", "12.45", "1", "15.6", "4", "25.5", "2", "4.30", "1"]
)
 ```

*Output:*

````python
#// BEGIN_TODO [string_list_simplified_splitser] Convert single string list into parameters for simplified_splitser

def string_list_simplified_splitser(data_list: list[str]):
    """Converts a list of strings with the amount of people, the costs made by each individual followed by its ID into a list of final balances of each individual.
    
    :param data_list: a list of the amount of people, strings of purchases followed by ID's of the person
    :type data_list: list(str)
    :rtype: str
    :return: a list of ID's and the final balances of every person in the household

    >>> string_list_simplified_splitser(["5", "12.45", "1", "15.6", "4", "25.5", "2", "4.30", "1"])
    0: € -11.57 
    1: € 5.18 
    2: € 13.93 
    3: € -11.57 
    4: € 4.03 

    >>> string_list_simplified_splitser(["5", "16.45", "1", "12.6", "4", "3.5", "2", "4.60", "2"])
    0: € -7.43
    1: € 9.02
    2: € 0.67
    3: € -7.43
    4: € 5.17

    >>> string_list_simplified_splitser(["5", "16.45", "1", "12.6", "4", "3.5", "2", "4.60"])
    Not every cost has an associated ID, please double-check every cost has an ID in the input list.

    """
    converted_list = []

    # when list is even, the converted list becomes uneven, because the first index is always the amount of people
    if len(data_list) % 2 == 0:
        raise Exception("Not every cost has an associated ID, please double-check every cost has an ID in the input list.")
    
    amount = int(data_list[0])

    for i in range(1, len(data_list), 2):
        if data_list[i].replace('.', '', 1).isdigit():
            costs = float(data_list[i])
            converted_list.append(costs)
        else: 
            raise Exception("Costs should be float, please check your input.")
        if data_list[i+1].isdigit():
            person = int(data_list[i+1])
            converted_list.append(person)
            if (person + 1) > amount:
                raise Exception("The ID is not in range with the amount of people in the household.")
        else:
            raise Exception("ID's should be ints, please check your input.")
        
    if len(converted_list) % 2 != 0:
        raise Exception("Either one ID is not an int or one cost is not a float, please check your input.")
        
    final_balances = simplified_splitser(amount, converted_list)

    for person, balance in enumerate(final_balances, start=0):
        print(f'{person}: € {round(balance, 2)}')
        
#// END_TODO [string_list_simplified_splitser]


string_list_simplified_splitser(
    ["5", "12.45", "1", "15.6", "4", "25.5", "2", "4.30", "1"]
)
````

## Task 4: Process Text File

Since we have created the function `string_list_simplified_splitser`, we are now ready to load the file `assets/simplified_splitser.txt` containing all the expenses of the household.

In order to do this, we now write a function `process_text_file` that 
- Opens the text file whose file name is given by `FILE_NAME`,
- reads it line by line and strips any leading and trailing whitespace from all its lines,
- transforms each line into a list of strings by splitting based on whitespaces
    - calls `string_list_simplified_splitser` on each household data (line) (this function prints) and 
    - **prints** *one* blank line in between

**Examples:**

*Input:*
<pre>
1 0.43 0
2 1.14 1 1.42 0
</pre>

*Output:* 

<pre>
0: € 0.00

0: € 0.14
1: € -0.14
</pre>

````python
FILE_NAME = 'assets/simplified_splitser.txt'

#// BEGIN_TODO [process_file] Process household expenses in text file 

def process_text_file(FILE_NAME: str) -> str:
    """Processes a text file with filename FILENAME and returns a list of ID's and final balances of the persons in the household.

    :param FILE_NAME: the name of the text file containing the input data
    :type FILE_NAME: str
    :rtype: str
    :return: a list of ID's and the final balances of every person in the household

    """

    with open(FILE_NAME) as f:
        while True:
            line = f.readline()
            if not line:
                break
            splitted_list = line.strip().split(' ')
            try:
                string_list_simplified_splitser(splitted_list)
            except Exception as e:
                print(e)
            print('')

#// END_TODO [process_file] 

process_text_file(FILE_NAME)
````

````python
#// BEGIN_TODO [bar_chart] Drawing a bar chart

def bar_chart(data):
    
    fig = plt.figure(figsize = (10, 5))

    ids = [i for i in range(0, len(data))]
    balances = [i for i in data]

    colors = ['orange' if b < 0 else 'blue' for b in balances]
    
    # creating the bar plot
    plt.bar(ids, balances, color =colors, width = 0.4)
    
    plt.xlabel("ID Group Members")
    plt.ylabel("Balances")
    plt.title("Final Balances")
    plt.show()

#// END_TODO [bar_chart]

data = simplified_splitser(5, [12.45, 1, 15.6, 4, 25.5, 2, 4.30, 1])
bar_chart(data)
````
