# Linear Least Squares

One of the main aspects of data analysis is to fit your data to analytical functions. In this exercise, we want to find a linear curve of the form $y=ax+b$ that fits best to the provided data. By solving for $a$ and $b$, you can find the optimal parameters that minimize the difference between the fit and the data. 

The difference between the fit and the data is determined by the residual. The data is given by a set of coordinates ($x_i, y_i$). The residual for the linear function $f(x) = a x + b $  is 
$$  
 R = \sum_{i=1}^N (y_i - f(x_i))^2 = \sum_{i=1}^N (y_i - (a x_i + b))^2. 
$$
To minimize the residual and find the best fit, one needs to minimize $R$ with respect to $a, b$, i.e. $\frac{\partial R}{\partial a} = 0 $ and $ \frac{\partial R}{\partial b} = 0 $ and solve the equations.
